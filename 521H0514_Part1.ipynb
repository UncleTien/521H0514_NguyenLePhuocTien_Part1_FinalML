```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Hàm loss function đơn giản
def simple_loss_function(x):
    return x**2

# Hàm loss function phức tạp
def complex_loss_function(x):
    return np.sin(x) + np.cos(x)

# Gradient descent (GD)
def gradient_descent(gradient_fn, lr, num_iterations):
    x = 10.0  # Giá trị khởi tạo
    losses = []

    for i in range(num_iterations):
        grad = gradient_fn(x)
        x -= lr * grad
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Batch gradient descent (BGD)
def batch_gradient_descent(gradient_fn, lr, num_iterations, batch_size=32):
    x = 10.0  # Giá trị khởi tạo
    losses = []

    for i in range(num_iterations):
        batch_x = np.random.uniform(-10, 10, size=batch_size)
        grad = np.mean([gradient_fn(xi) for xi in batch_x])
        x -= lr * grad
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Stochastic Gradient Descent (SGD)
def stochastic_gradient_descent(gradient_fn, lr, num_iterations):
    x = 10.0  # Giá trị khởi tạo
    losses = []

    for i in range(num_iterations):
        idx = np.random.randint(-10, 10)  # Chọn ngẫu nhiên một điểm
        grad = gradient_fn(idx)
        x -= lr * grad
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Mini-batch gradient descent
def mini_batch_gradient_descent(gradient_fn, lr, num_iterations, batch_size=32):
    x = 10.0  # Giá trị khởi tạo
    losses = []

    for i in range(num_iterations):
        batch_x = np.random.uniform(-10, 10, size=batch_size)
        grad = np.mean([gradient_fn(xi) for xi in batch_x])
        x -= lr * grad
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Momentum
def momentum(gradient_fn, lr, num_iterations, gamma=0.9):
    x = 10.0  # Giá trị khởi tạo
    velocity = 0
    losses = []

    for i in range(num_iterations):
        grad = gradient_fn(x)
        velocity = gamma * velocity + lr * grad
        x -= velocity
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# RMSprop
def rmsprop(gradient_fn, lr, num_iterations, beta=0.9, epsilon=1e-8):
    x = 10.0  # Giá trị khởi tạo
    cache = 0
    losses = []

    for i in range(num_iterations):
        grad = gradient_fn(x)
        cache = beta * cache + (1 - beta) * grad**2
        x -= lr * grad / (np.sqrt(cache) + epsilon)
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Adam
def adam(gradient_fn, lr, num_iterations, beta1=0.9, beta2=0.999, epsilon=1e-8):
    x = 10.0  # Giá trị khởi tạo
    m = 0
    v = 0
    losses = []

    for i in range(num_iterations):
        grad = gradient_fn(x)
        m = beta1 * m + (1 - beta1) * grad
        v = beta2 * v + (1 - beta2) * grad**2
        m_hat = m / (1 - beta1**(i + 1))
        v_hat = v / (1 - beta2**(i + 1))
        x -= lr * m_hat / (np.sqrt(v_hat) + epsilon)
        loss = simple_loss_function(x)  # Tính loss
        losses.append(loss)

    return losses

# Đo độ dốc của hàm loss function đơn giản
def gradient_simple_loss(x):
    return 2 * x

# Đo độ dốc của hàm loss function phức tạp
def gradient_complex_loss(x):
    return np.cos(x) - np.sin(x)

# Số lần lặp và learning rate
num_iterations = 100
learning_rate = 0.1

```

```python
# Chạy các thuật toán tối ưu trên hàm loss function đơn giản và phức tạp
gd_losses = gradient_descent(gradient_simple_loss, learning_rate, num_iterations)
bgd_losses = batch_gradient_descent(gradient_simple_loss, learning_rate, num_iterations)
sgd_losses = stochastic_gradient_descent(gradient_simple_loss, learning_rate, num_iterations)
mbgd_losses = mini_batch_gradient_descent(gradient_simple_loss, learning_rate, num_iterations)
momentum_losses = momentum(gradient_simple_loss, learning_rate, num_iterations)
rmsprop_losses = rmsprop(gradient_simple_loss, learning_rate, num_iterations)
adam_losses = adam(gradient_simple_loss, learning_rate, num_iterations)

# Vẽ biểu đồ so sánh trên hàm loss function đơn giản
plt.figure(figsize=(10, 6))
plt.plot(gd_losses, label='Gradient Descent')
plt.plot(bgd_losses, label='Batch Gradient Descent')
plt.plot(sgd_losses, label='SGD')
plt.plot(mbgd_losses, label='Mini-Batch GD')
plt.plot(momentum_losses, label='Momentum')
plt.plot(rmsprop_losses, label='RMSprop')
plt.plot(adam_losses, label='Adam')
plt.title('Comparison of Optimizers on Simple Loss Function')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Chạy các thuật toán tối ưu trên hàm loss function phức tạp
gd_losses_complex = gradient_descent(gradient_complex_loss, learning_rate, num_iterations)
bgd_losses_complex = batch_gradient_descent(gradient_complex_loss, learning_rate, num_iterations)
sgd_losses_complex = stochastic_gradient_descent(gradient_complex_loss, learning_rate, num_iterations)
mbgd_losses_complex = mini_batch_gradient_descent(gradient_complex_loss, learning_rate, num_iterations)
momentum_losses_complex = momentum(gradient_complex_loss, learning_rate, num_iterations)
rmsprop_losses_complex = rmsprop(gradient_complex_loss, learning_rate, num_iterations)
adam_losses_complex = adam(gradient_complex_loss, learning_rate, num_iterations)

# Vẽ biểu đồ so sánh trên hàm loss function phức tạp
plt.figure(figsize=(10, 6))
plt.plot(gd_losses_complex, label='Gradient Descent')
plt.plot(bgd_losses_complex, label='Batch Gradient Descent')
plt.plot(sgd_losses_complex, label='SGD')
plt.plot(mbgd_losses_complex, label='Mini-Batch GD')
plt.plot(momentum_losses_complex, label='Momentum')
plt.plot(rmsprop_losses_complex, label='RMSprop')
plt.plot(adam_losses_complex, label='Adam')
plt.title('Comparison of Optimizers on Complex Loss Function')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
